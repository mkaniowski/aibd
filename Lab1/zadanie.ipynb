## Zadanie 3

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def fun1(x: int):
    return x*x + 5
```

```python
# wyznaczam kolejne argumenty funkcji
x1 = np.linspace(-6, 6, 100)
x2 = np.linspace(-1, 1, 100)
x3 = np.linspace(0, 5, 100)

# plot funkcji
plt.plot(x1, fun1(x1), linewidth='4', label='-6<x<6')
plt.plot(x2, fun1(x2), linewidth='3', label='-1<x<1')
plt.plot(x3, fun1(x3), linewidth='2', label='0<x<5')

plt.title('x^2 + 5')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```
