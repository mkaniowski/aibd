```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Original Data/earthquake_data.csv')
```

```python
data
```

```python
# data.info()
# data.describe()
```

```python
data = data.rename(columns={"What is your gender?": "sex", "Age": "age", "How worried are you about the Big One, a massive, catastrophic earthquake?": "bigOne", """Do you think the "Big One" will occur in your lifetime?""": "bigOneOccurrence", "How much total combined money did all members of your HOUSEHOLD earn last year?": "income", "Have you or anyone in your household taken any precautions for an earthquake (packed an earthquake survival kit, prepared an evacuation plan, etc.)?":"precautions", "In general, how worried are you about earthquakes?":"worry"})
```

```python
data
```

```python
analysis = data.loc[:, ["sex", "age", "bigOne"]]
analysis.dropna(how='any', inplace=True)
analysis.to_csv('../Analysis data/bigOne.csv')
```

```python
fig=plt.figure(figsize=(10, 5))
sns.countplot(data, x="bigOne", hue="sex")
plt.title("How worry is the person about the Big One based on sex")
plt.grid()
plt.show()
```

```python
fig=plt.figure(figsize=(20, 5))
sns.countplot(data, x="income", hue="bigOneOccurrence")
plt.title("Belif in the Big One occurring in the lifetime based on the income")
plt.grid()
plt.show()
```

```python
fig=plt.figure(figsize=(4, 4))
sns.countplot(data, x="bigOneOccurrence", hue="precautions")
plt.title("Any taken precautions based on belif in big one occurring in the lifetime")
plt.grid()
plt.show()
```

```python
fig=plt.figure(figsize=(10, 4))
sns.countplot(data, x="worry", hue="bigOne")
plt.title("How worry are people of the Big one based on their generaly worry about erathquakes")
plt.grid()
plt.show()
```
